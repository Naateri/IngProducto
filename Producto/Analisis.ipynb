```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
preguntas = ['Time', 'Edad', 'Sexo', 'Importancia', 'Informar', 'Revisar_UV', 'Consciente', 'Reacciones', 'Uso_Bloqueador',
            'Uso_Gorro', 'Uso_Lentes', 'Protege', 'Productos', 'PorQueNo']

edades = ["De 15 a 20 años", "De 21 a 30 años", "De 31 a 40 años", "De 41 a 50 años", "De 51 a más años"]

# Informar: cree que se informa lo suficiente en aqp
# Consciente: es usted consciente 
encuestas = pd.read_csv('Radiacion.csv', sep=',')
encuestas.columns = preguntas
encuestas
```

```python
e_hombre = encuestas[encuestas["Sexo"] == "Masculino"]
e_mujer = encuestas[encuestas["Sexo"] == "Femenino"]
```

```python
hombre_importancia = e_hombre["Importancia"].value_counts()
mujer_importancia = e_mujer["Importancia"].value_counts()

hombre_importancia_pct = [0] * len(np.array(hombre_importancia))
mujer_importancia_pct = [0] * len(np.array(mujer_importancia))

for i in range(len(np.array(hombre_importancia))):
    hombre_importancia_pct[i] = np.array(hombre_importancia)[i] / sum(np.array(hombre_importancia)) * 100
    
for i in range(len(np.array(mujer_importancia))):
    mujer_importancia_pct[i] = np.array(mujer_importancia)[i] / sum(np.array(mujer_importancia)) * 100

```

```python
# np.array(hombre_importancia) # 5, 4, 3
plt.title('Importancia protección rayos UV - Hombres')
plt.bar(['3','4','5'], np.array(hombre_importancia)[::-1])
plt.show()
```

```python
plt.title('Importancia protección rayos UV - Hombres (%)')
plt.pie(np.array(hombre_importancia)[::-1], labels=['3', '4', '5'], autopct='%1.2f%%')
plt.savefig('Importancia_Hombres.png')
plt.show()
```

```python
plt.title('Importancia protección rayos UV - Mujeres')
plt.bar(['1', '3','4','5'], np.array(mujer_importancia)[::-1], color='red')
plt.show()
```

```python
plt.title('Importancia protección rayos UV - Mujeres (%)')
plt.pie(np.array(mujer_importancia)[::-1], labels=['1', '3', '4', '5'], autopct='%1.2f%%')
plt.savefig('Importancia_Mujeres.png')
plt.show()
```

```python
hombre_productos = np.array(e_hombre["Productos"])
total_hproductos = 0
h_combos = [0] * 4 
# ej: [0,4,5,1] => 0 usan solo 1, 4 usan 2, 5 usan 3 y 1 usa 4

productos_hombres = dict()
for producto in hombre_productos:
    if type(producto) != str:
        continue
    total_hproductos += 1
    cur_productos = producto.split(';') # list of products
    
    h_combos[len(cur_productos)-1] += 1
    
    for producto in cur_productos:
        if producto in productos_hombres:
            productos_hombres[producto] += 1
        else:
            productos_hombres[producto] = 1
    
print(productos_hombres)
print(h_combos)
print(total_hproductos)
```

```python
prod_hombres_pct = productos_hombres.copy()
for key in prod_hombres_pct.keys():
    prod_hombres_pct[key] = productos_hombres[key] / total_hproductos * 100

print(prod_hombres_pct)

plt.title('Productos utilizados - Hombres (%)')
plt.bar(range(len(prod_hombres_pct)), list(prod_hombres_pct.values()), align='center')
plt.xticks(range(len(prod_hombres_pct)), list(prod_hombres_pct.keys()))
plt.savefig('Productos Usados - Hombres(%).png')
plt.show()
```

```python
plt.title('Cantidad de productos utilizados - Hombres')
plt.bar(['1', '2', '3', '4'], h_combos)
plt.savefig('Productos usados - Combinación - Hombres.png')
plt.show()
```

```python
mujer_productos = np.array(e_mujer["Productos"])
total_mproductos = 0
m_combos = [0] * 4 
# ej: [0,4,5,1] => 0 usan solo 1, 4 usan 2, 5 usan 3 y 1 usa 4

productos_mujeres = dict()
for producto in mujer_productos:
    if type(producto) != str:
        continue
    total_mproductos += 1
    cur_productos = producto.split(';') # list of products
    
    m_combos[len(cur_productos)-1] += 1
    
    for producto in cur_productos:
        if producto in productos_mujeres:
            productos_mujeres[producto] += 1
        else:
            productos_mujeres[producto] = 1
    
print(productos_mujeres)
print(m_combos)
print(total_mproductos)
```

```python
prod_mujeres_pct = productos_mujeres.copy()
for key in prod_mujeres_pct.keys():
    prod_mujeres_pct[key] = productos_mujeres[key] / total_mproductos * 100

print(prod_mujeres_pct)


plt.title('Productos utilizados - Mujeres (%)')
plt.bar(range(len(prod_mujeres_pct)), list(prod_mujeres_pct.values()), align='center', color='red')
plt.xticks(range(len(prod_mujeres_pct)), list(prod_mujeres_pct.keys()), rotation=30)
plt.savefig('Productos Usados - Mujeres(%).png')
plt.show()
```

```python
plt.title('Cantidad de productos utilizados - Mujeres')
plt.bar(['1', '2', '3', '4'], m_combos, color='red')
plt.savefig('Productos usados - Combinación - Mujeres.png')
plt.show()
```
